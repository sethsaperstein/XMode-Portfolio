```python
header_names = ["app_id", "ad_id", 'platform', 'latitude', 'longitude', 'horizontal_accuracy', 'location_at', 'email', 'ipv_4', 'user_agent', 'country', 'battery', 'background', 'network', 'venue_name', 'venue_category', 'vertical_accuracy', 'speed', 'heading', 'publisher_id', 'dwell_time', 'person_id', 'client_id', 'sdk_version', 'altitude', 'ipv_6', 'mobile_number', 'captured_at', 'created_at', 'floor', 'carrier', 'manufacturer', 'device_model', 'wifi_ssid', 'wifi_bssid', 'decorated_at', 'dist_moved', 'day_number', 'day_type', 'time_type', 'dwell_type', 'confidence', 'brand_name', 'source', 'tech_signals', 'geo_country', 'misc']
```

```python
POINTS_LIMIT = 5 #minutes
TIME_LIMIT = '1 min'
```

```python
def average_distance(merged):
  """
  merged: a table with lats and lons for sdk and s2s data at the same timestamp
  returns a list of distances
  """
  distances = []
  bad = []
  for i in range(len(merged)):
    lat1 = merged.iloc[i, 1]
    lon1 = merged.iloc[i, 2]
    lat2 = merged.iloc[i, 4]
    lon2 = merged.iloc[i, 5]
    dist = distance.distance((lat1, lon1), (lat2, lon2)).m
    distances.append(dist)
    if dist > 100:
      bad.append(i)
  return distances, bad
```

```python
pub_ids = ["areametrics_id_overlap"]
ad_ids = ["android_id_overlap"]
```

```python
s2s = pd.read_csv(pub_ids[0] + ".csv", names=header_names)
sdk = pd.read_csv(ad_ids[0] + ".csv", names=header_names)
s2s['location_at'] = pd.to_datetime(s2s['location_at'])
s2s.dropna(subset=['location_at'], inplace=True)
sdk.dropna(subset=['location_at'], inplace=True)
sdk['location_at'] = pd.to_datetime(sdk['location_at'])
sdk.sort_values("location_at", inplace=True)
s2s.sort_values("location_at", inplace=True)
merged = pd.merge_asof(sdk, s2s, on='location_at', by='ad_id', tolerance=pd.Timedelta(TIME_LIMIT))
df = merged[["ad_id", "latitude_x", "longitude_x", "location_at", "latitude_y", "longitude_y"]].dropna()
```

```python
avgs = []
for pub_id in pub_ids:
  s2s = pd.read_csv(pub_id + ".csv", names=header_names)
  for ad_id in ad_ids:
    sdk = pd.read_csv(ad_id + ".csv", names=header_names)
    merged = pd.merge_asof(sdk, s2s, on='location_at', by='ad_id', tolerance=pd.Timedelta(TIME_LIMIT), direction='nearest')
    if merged.shape[0] < POINTS_LIMIT:
      continue
    avgs.extend(average_distance(merged))
print(sum(avgs) / len(avgs))
```
